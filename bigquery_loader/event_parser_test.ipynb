## Event Data Descriptions
- `id`: game identifier
- `info`: bunch of game metadata like when and where the game is and who are the umpires among other things
- `version`: can be ignored
- `start`: starting lineups for teams
- `sub`: when players are substituted
- `play`: results of at-bats (can also have no-play [NP] prior to substitutions)
- `badj`: for when a batter hits from the opposite side of the plate than is expected
- `padj`: for when a pitcher throws from the opposite than is expected (only two such players)
- `ladj`: for when a team bats out of order
- `radj`: for extra-innings runner entries starting in 2020
- `presadj`: some weird stuff around pitcher responsibility
- `data`: pretty much just earned runs for pitchers
- `com`: commentary for some events (like ML debut for player or ejections)

```python
class EventLog:
    def __init__(self, records):
        self.info = {}
        self.plays = []
        self.data = []
        play_cnt = 0
        for record in records:
            if record[0] == 'id':
                self.id = record[1]
                self.lineups = {
                    # lineup key will be <gameid>-<home/away>-<version>
                    f'{self.id}-0-0': [],
                    f'{self.id}-1-0': [],
                }
                home_lineup = {
                    'team': 0,
                    'version': 0,
                    'players': [],
                }
                away_lineup = {
                    'team': 0,
                    'version': 0,
                    'players': [],
                }
            elif record[0] == 'info':
                self.info[record[1]] = record[2]
            elif record[0] == 'start':
                temp_dict = {}
                temp_dict['player_id'] = record[1]
                temp_dict['player_name'] = record[2]
                temp_dict['batting_order'] = record[4]
                temp_dict['position'] = record[5]
                self.lineups[f'{self.id}-{record[3]}-0'].append(temp_dict)
                if record[3] == '1':
                    home_lineup['players'].append(temp_dict)
                else:
                    away_lineup['players'].append(temp_dict)
            elif record[0] == 'play':
                temp_dict = {}
                temp_dict['inning'] = record[1]
                temp_dict['inning_half'] = 'top' if record[2] == '0' else 'bottom'
                temp_dict['player_id'] = record[3]
                temp_dict['count'] = record[4]
                temp_dict['pitch_info'] = record[5]
                temp_dict['result'] = record[6]
                temp_dict['play_cnt'] = play_cnt
                temp_dict['away_lineup'] = f'{self.id}-0-{away_lineup["version"]}'
                temp_dict['home_lineup'] = f'{self.id}-1-{home_lineup["version"]}'
                self.plays.append(temp_dict)
                play_cnt += 1
            elif record[0] == 'sub':
                temp_dict = {}
                temp_dict['player_id'] = record[1]
                temp_dict['player_name'] = record[2]
                temp_dict['batting_order'] = record[4]
                temp_dict['position'] = record[5]
                if record[3] == '1':
                    # increment version
                    home_lineup['version'] += 1
                    # remove player in that spot in batting order
                    home_lineup['players'] = [player for player in home_lineup['players']if player['batting_order'] != temp_dict['batting_order']]
                    # add new player to lineup
                    home_lineup['players'].append(temp_dict)
                    # add new entry to master lineup
                    self.lineups[f'{self.id}-{record[3]}-{home_lineup["version"]}'] = home_lineup['players']
                else:
                    # increment version
                    away_lineup['version'] += 1
                    # remove player in that spot in batting order
                    away_lineup['players'] = [player for player in away_lineup['players']if player['batting_order'] != temp_dict['batting_order']]
                    # add new player to lineup
                    away_lineup['players'].append(temp_dict)
                    # add new entry to master lineup
                    self.lineups[f'{self.id}-{record[3]}-{away_lineup["version"]}'] = away_lineup['players']
            elif record[0] == 'data':
                self.data.append(record[1:])

    def get_game_lineup_table(self):
        pd_list = []
        for k, v in self.lineups.items():
            game_id, team, version = k.split('-')
            df_tmp = pd.DataFrame(v)
            df_tmp['key'] = k
            df_tmp['game_id'] = game_id
            df_tmp['home_away'] = team
            df_tmp['version'] = version
            pd_list.append(df_tmp)
        return (pd.concat(pd_list))
    
    def get_plays_table(self):
        df_plays = pd.DataFrame(self.plays)
        df_plays['count'] = df_plays['count'].apply(lambda x: f'{x[0]}-{x[1]}')
        df_plays['result_basic'] = df_plays['result'].apply(lambda x: x.split('.')[0].split('/')[0])
        return (df_plays)

    def get_batting_stats():
        # Create player-game level batting stats
        pass

    def get_pitching_stats():
        # Create player-game level pitching stats
        pass
```
